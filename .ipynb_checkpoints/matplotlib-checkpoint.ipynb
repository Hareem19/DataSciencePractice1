```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
s2 = pd.Series([12,24,36,12,6,12,24], index=['apples','oranges','bananas', 'samosas','rolls', 'chickens', 'coldrinks']) 
s2
```

```python
plt.bar(s2.index,s2.values)
plt.xlabel("fruits")
plt.xlabel("nums")
plt.grid(True)
plt.savefig("myfirstplot")
# plt.legend()
plt.show()
plt.show()
```
